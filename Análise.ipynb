```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
from pyUFbr.baseuf import ufbr
```

```python
# Carregar arquivos
df_2024 = pd.read_csv('dados/resultado_2024.csv', delimiter=',', encoding='utf-8')
df_2023 = pd.read_csv('dados/resultado_2023.csv', delimiter=',', encoding='utf-8')
df_2022 = pd.read_csv('dados/resultado_2022.csv', delimiter=',', encoding='utf-8')
df_2021 = pd.read_csv('dados/resultado_2021.csv', delimiter=',', encoding='utf-8')
df_2020 = pd.read_csv('dados/resultado_2020.csv', delimiter=',', encoding='utf-8')
```

```python
dfs = [(2020, df_2020),(2021, df_2021),(2022, df_2022), (2023, df_2023), (2024, df_2024)]

for ano, df in dfs:

    print(f"\n\nAnalise do df: {ano}")

    print(f"\nValores Ausentes:")
    print(df.isna().sum())
    
    try: 
        df['data_inversa'] = pd.to_datetime(df['data_inversa'])
        df['horario'] = pd.to_datetime(df['horario'], format='%H:%M:%S').dt.time

        valores_anos = df['data_inversa'].dt.year.value_counts().keys()
        if list(valores_anos) == [df['data_inversa'].dt.year[0]] and len(valores_anos) == 1:
            print("\nData e horário estão certos!")
        else:
            print("\nExiste um valor do ano que não corresponde a tabela")
    except: 
        print("\nData ou horário possuem erros!")
    
    print("\nVerificação dos valores dia da semana:")
    print(sorted(df['dia_semana'].value_counts().keys()))
    
    print("\nVerificação dos valores fase do dia:")
    valores_fase_dia = sorted(df['fase_dia'].value_counts().keys())
    print(valores_fase_dia)

    for valor in valores_fase_dia:
        df_fase = df[df['fase_dia'] == valor]
        
        horarios_decimal = df_fase['horario'].apply(lambda x: x.hour + x.minute / 60.0)
        
        plt.hist(horarios_decimal, bins=24, range=(0, 24), color='skyblue', edgecolor='black', alpha=0.7)
        
        plt.xlabel('Horário (em horas)')
        plt.ylabel('Número de valores')
        plt.title(f'Distribuição de horários para a fase {valor}')
        
        plt.xticks(range(0, 25), [f"{h:02d}:00" for h in range(24)] + ["23:59"], rotation=90)
        
        plt.tight_layout()
        plt.show()

    print("\nVerificação dos valores do estado:")
    # Verifica se os estado existe
    df['estado_valido'] = df['uf'].apply(lambda sigla: sigla.upper() in ufbr.list_uf)
    
    # Verifica se existe algum valor False na coluna 'estado_valido'
    estado_invalido = not df['estado_valido'].all()
    
    if estado_invalido:
        print("Existe pelo menos 1 estado inválido.")
        print(df[['estado_valido']==False])
    else:
        print("Todos os estados são validos.")

    print("\nVerificação se existe mais pessoas feridas do que pessoas numa linha:")
    df_verificacao = df.query('pessoas < feridos')
    
    if not df_verificacao.empty:
        print("Existem casos onde o número de pessoas feridas é maior que o número de pessoas.")
        print(df_verificacao)
    else:
        print("Não há casos onde o número de pessoas feridas é maior que o número de pessoas.")


    print("\nVerificação se existe mais mortos do que pessoas numa linha:")
    df_verificacao = df.query('pessoas < mortos')
    
    if not df_verificacao.empty:
        print("Existem casos onde o número de mortos é maior que o número de pessoas.")
        print(df_verificacao[['pessoas', 'mortos']])
    else:
        print("Não há casos onde o número mortos é maior que o número de pessoas.")

    print("\nVerificação dos valores pessoas envolvidas:")
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

    # Histograma
    ax1.hist(df['pessoas'], color='skyblue', edgecolor='black', alpha=0.7)
    ax1.set_xlabel('Pessoas envolvidas')
    ax1.set_ylabel('Número de valores')
    ax1.set_title('Distribuição de pessoas envolvidas')
    
    # Boxplot
    ax2.boxplot(df['pessoas'], widths=0.5, patch_artist=True, boxprops=dict(facecolor='skyblue'))
    ax2.set_xlabel('Pessoas envolvidas')
    ax2.set_title('Boxplot de pessoas envolvidas')
    
    plt.tight_layout()
    plt.show()
    
    print("\nVerificação dos valores pessoas feridas:")
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
    
    # Histograma
    ax1.hist(df['feridos'], color='skyblue', edgecolor='black', alpha=0.7)
    ax1.set_xlabel('Pessoas feridas')
    ax1.set_ylabel('Número de valores')
    ax1.set_title('Distribuição de pessoas feridas')
    
    # Boxplot
    ax2.boxplot(df['feridos'], widths=0.5, patch_artist=True, boxprops=dict(facecolor='skyblue'))
    ax2.set_xlabel('Pessoas feridas')
    ax2.set_title('Boxplot de pessoas feridas')
    
    plt.tight_layout()
    plt.show()
    
    print("\nVerificação dos valores de mortos:")
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

    # Histograma
    ax1.hist(df['mortos'], color='skyblue', edgecolor='black', alpha=0.7)
    ax1.set_xlabel('Quantidade mortos')
    ax1.set_ylabel('Número de valores')
    ax1.set_title('Distribuição de mortos')
    
    # Boxplot
    ax2.boxplot(df['mortos'], widths=0.5, patch_artist=True, boxprops=dict(facecolor='skyblue'))
    ax2.set_xlabel('Mortos')
    ax2.set_title('Boxplot de mortos')
    
    plt.tight_layout()
    plt.show()    
    
    print("\nVerificação dos valores de número de veiculos:")
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

    # Histograma
    ax1.hist(df['veiculos'], color='skyblue', edgecolor='black', alpha=0.7)
    ax1.set_xlabel('Quantidade veículos')
    ax1.set_ylabel('Número de valores')
    ax1.set_title('Distribuição de veículos')
    
    # Boxplot
    ax2.boxplot(df['veiculos'], widths=0.5, patch_artist=True, boxprops=dict(facecolor='skyblue'))
    ax2.set_xlabel('Veículos')
    ax2.set_title('Boxplot de Veículos')
    
    plt.tight_layout()
    plt.show()    

    print("\nVerificação se os valores de pessoas for 0 (nulo):")
    pessoas_nulo = df[df['pessoas'] == 0]
    if pessoas_nulo.empty:
        print("Não há valores 0 em pessoas")
    else:
        print("Há valores 0 em pessoas")

    print("\nVerificação se os valores de veículo for 0 (nulo):")
    veiculos_nulo = df[df['veiculos'] == 0]
    if veiculos_nulo.empty:
        print("Não há valores 0 em veículo")
    else:
        print("Há valores 0 em veículo")

    print("\nVerificação dos valores do sentido da via:")
    print(sorted(df['sentido_via'].value_counts().keys()))

    print("\nVerificação dos valores do tipo da pista:")
    print(sorted(df['tipo_pista'].value_counts().keys()))

    print("\nVerificação dos valores do uso do solo:")
    print(sorted(df['uso_solo'].value_counts().keys()))

    print("\nVerificação dos valores do tracado da via:")
    print(sorted(df['tracado_via'].value_counts().keys()))

    print("\nVerificação dos valores da classificação do acidente:")
    print(sorted(df['classificacao_acidente'].value_counts().keys()))

    print("\nVerificação dos valores do tipo do acidente:")
    print(sorted(df['tipo_acidente'].value_counts().keys()))

    print("\nVerificação dos valores da condição metereologica:")
    print(sorted(df['condicao_metereologica'].value_counts().keys()))

    print("\nVerificação dos valores da causa do acidente:")
    print(sorted(df['causa_acidente'].value_counts().keys()))

    print("\nVerificação dos valores do tipo do veículo:")
    print(sorted(df['tipo_veiculo'].value_counts().keys()))

    print("\nVerificação dos valores do ano de fabricação do veículo:")
    print(sorted(df['ano_fabricacao_veiculo'].value_counts().keys()))

    print("\nVerificação do número de valores do ano de fabricação do veículo iguais a 0:")
    print(sorted(df[df['ano_fabricacao_veiculo']==0]['ano_fabricacao_veiculo'].value_counts()))
```
